## Part 1

```python
test_puzzle_input_part_1 = """Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11
""".splitlines()
```

```python
def part_1(input):
    points = 0

    for line in input:
        parts = line.split(':')
        number_parts = parts[1].split('|')
        winning_numbers = number_parts[0].split()
        current_numbers = number_parts[1].split()
        
        card_points = 0

        for winning_number in winning_numbers:
            if winning_number in current_numbers:
                card_points = max(1, card_points * 2)

        points = points + card_points

    return points

assert part_1(test_puzzle_input_part_1) == 13
```

```python
part_1(puzzle_input)
```

## Part 2--- Part Two ---

Just as you're about to report your findings to the Elf, one of you realizes that the rules have actually been printed on the back of every card this whole time.

There's no such thing as "points". Instead, scratchcards only cause you to win more scratchcards equal to the number of winning numbers you have.

Specifically, you win copies of the scratchcards below the winning card equal to the number of matches. So, if card 10 were to have 5 matching numbers, you would win one copy each of cards 11, 12, 13, 14, and 15.

Copies of scratchcards are scored like normal scratchcards and have the same card number as the card they copied. So, if you win a copy of card 10 and it has 5 matching numbers, it would then win a copy of the same cards that the original card 10 won: cards 11, 12, 13, 14, and 15. This process repeats until none of the copies cause you to win any more cards. (Cards will never make you copy a card past the end of the table.)

This time, the above example goes differently:

```python
test_puzzle_input_part_2 = """Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11
""".splitlines()
```

```python
class Card:
    def __init__(self, line) -> None:
        parts = line.split(':')
        number_parts = parts[1].split('|')
        self.id = int(parts[0].replace('Card ', ''))
        self.winning_numbers = number_parts[0].split()
        self.current_numbers = number_parts[1].split()
        self.matching_numbers = self.get_matching_numbers()

    def __str__(self) -> str:
        return f'Card {self.id}'
    
    def __repr__(self) -> str:
        return self.__str__()

    def get_matching_numbers(self):
        return [number for number in self.winning_numbers if number in self.current_numbers]

def get_winning_cards(card, cards, cards_counter):
    cards_counter[card.id] = cards_counter[card.id] + 1
    matching_numbers = card.get_matching_numbers()
    matching_numbers_count = len(matching_numbers)
    if(matching_numbers_count == 0):
        return
    
    winning_cards = cards[card.id:card.id + matching_numbers_count]
    
    for winning_card in winning_cards:
        get_winning_cards(winning_card, cards, cards_counter)


def part_2(input):
    cards_counter = {}
    cards = [Card(line) for line in input]
    
    for card in cards:
        cards_counter[card.id] = 0

    for card in cards:
        get_winning_cards(card, cards, cards_counter)
    

    return sum(list(cards_counter.values()))

assert part_2(test_puzzle_input_part_2) == 30
```

```python
part_2(puzzle_input)
```
