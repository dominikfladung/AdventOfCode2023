## Part 1

```python
test_puzzle_input_part_1 = """32T3K 765
T55J5 684
KK677 28
KTJJT 220
QQQJA 483
""".splitlines()
```

```python
labels = ['A','K','Q','J','T','9','8','7','6','5','4','3','2']
def is_five_of_a_kind(hand_index):
    return len(hand_index) == 1

def is_four_of_a_kind(hand_index):
    if len(hand_index) != 2:
        return False
    return 4 in hand_index.values()
    
def is_full_house(hand_index):
    if len(hand_index) != 2:
        return False

    return 3 in hand_index.values()

def is_three_of_a_kind(hand_index):
    if len(hand_index) != 3:
        return False

    return 3 in hand_index.values() and 1 in hand_index.values()

def is_tour_pair(hand_index):
    if len(hand_index) != 3:
        return False
    return list(hand_index.values()).count(2) == 2

def is_one_pair(hand_index):
    return 2 in hand_index.values()

def get_index(hand):
    return {x: hand.count(x) for x in set(hand)}

def score_labels(hand):
    labels_score = 0
    for i in range(len(hand)):
        labels_score += (len(labels) - labels.index(hand[i]) - 1) * 100**(len(hand) - i - 1)
    return labels_score

def score_hand(hand):
    hand_index = get_index(hand)
    if is_five_of_a_kind(hand_index):
        return 6
    if is_four_of_a_kind(hand_index):
        return 5
    if is_full_house(hand_index):
        return 4
    if is_three_of_a_kind(hand_index):
        return 3
    if is_tour_pair(hand_index):
        return 2
    if is_one_pair(hand_index):
        return 1
    return 0

def part_1(input):
    cards = [(x[:5], int(x[6:]), score_hand(x[:5]), score_labels(x[:5])) for x in input]
    sorted_cards = sorted(cards, key=lambda x: (x[2], x[3]))
    solution = 0
    for rank in range(len(sorted_cards)):
        solution += sorted_cards[rank][1] * (rank + 1)

    return solution

assert part_1(test_puzzle_input_part_1) == 6440
```

```python
part_1(puzzle_input)
```
